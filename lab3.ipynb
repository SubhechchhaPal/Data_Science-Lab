Autumn, Winter, Spring or Summer.


```python
def Season(month):
  Winter=["january","december","november"]
  Spring=["february","march","april"]
  Summer=["may","june","july"]
  Autumn=["august","september","october"]
  if month in Winter:
    print("Winter")
  elif month in Spring:
    print("Spring")
  elif month in Summer:
    print("Summer")
  elif month in Autumn:
    print("Autumn")
  else:
    print("Worng Month")

month=input(("Enter Month: ")).lower()
Season(month)
```

2. Write a function called calculate_slope which return the slope of a linear equation


```python
def calculate_slope(x1, y1, x2, y2):
  if x1 == x2:
    raise ValueError("The x-coordinates of the two points must be different to calculate a slope.")
  slope = (y2 - y1) / (x1 - x2)
  return slope

print("The coordinates of the first point: ")
x1= int(input("x1: "))
y1=int(input("y1: "))
print("The coordinates of the second point: ")
x2= int(input("x2: "))
y2=int(input("y2: "))
slope = calculate_slope(x1, y1, x2, y2)
print("The slope of the line through points", x1,y1, "and", x2,y2, "is",slope)
```

```python
import math
def solve_quadratic_eqn(A,B,C):
  delta=int(pow(B,2)-4*A*C)
  if delta<0:
    print("The equation has two complex roots, which are conjugates of each other.")
  else:
    x1=((-B+math.sqrt(delta))/(2*A))
    x2=((-B-math.sqrt(delta))/(2*A))
    equation_set=[x1,x2]
    print("Solution set is:",equation_set)

print("Enter coefficients of the a,b,c: ")
a= int(input("a: "))
b=int(input("b: "))
c= int(input("c: "))
solve_quadratic_eqn(a,b,c)
```

6. Compute the sum up to n terms in the series
1 - 1/2 + 1/3 - 1/4 + 1/5 -... 1/n where n is a positive integer and input by user.

```python
def alternating_harmonic_sum(n):
    sum = 0.0
    for i in range(1, n + 1):
        sum += (-1) ** (i + 1) / i
    return sum
n = int(input("Enter a positive integer: "))
result = alternating_harmonic_sum(n)
print(f"The sum of the series up to {n} terms is {result}")
```

7. Write a program to compute sin x for given x. The user should supply x and a positive integer
n. We compute the sine of x using the series and the computation should use all terms in the
series up through the term involving xn
sin x = x - x3/3! + x5/5! - x7/7! + x9/9! ........

```python
import math
def factorial(n):
    if n == 0:
        return 1
    else:
        return n * factorial(n-1)
def compute_sin(x, n):
    sin_x = 0
    for i in range(n):
        sign = (-1) ** i
        term = (x ** (2 * i + 1)) / factorial(2 * i + 1)
        sin_x += sign * term
    return sin_x
x = float(input("Enter the value of x (in radians): "))
n = int(input("Enter the number of terms n: "))
result = compute_sin(x, n)
print(f"The computed value of sin({x}) using {n} terms is: {result}")
print(f"The value of sin({x}) from math.sin() is: {math.sin(x)}")
```

8. Write a program to compute cosine of x. The user should supply x and a positive integer n.
We compute the cosine of x using the series and the computation should use all terms in the
series up through the term involving xn
cos x = 1 - x2/2! + x4/4! - x6/6! ....

```python
import math

def factorial(n):
    if n == 0:
        return 1
    else:
        return n * factorial(n-1)
def compute_cos(x, n):
    cos_x = 0
    for i in range(n):
        sign = (-1) ** i
        term = (x ** (2 * i)) / factorial(2 * i)
        cos_x += sign * term
    return cos_x
x = float(input("Enter the value of x (in radians): "))
n = int(input("Enter the number of terms n: "))
result = compute_cos(x, n)
print(f"The computed value of cos({x}) using {n} terms is: {result}")
print(f"The value of cos({x}) from math.cos() is: {math.cos(x)}")
```

9. Print the pattern upto N Lines:

.
/_\
.
/ \
/___\
.
/ \
/ \
/_____\

N=2 N=3 N=4

```python
def print_pattern(N):
    print(".")
    for i in range(1, N ):
        for j in range(N - i):
            print(end="")
        print("/", end="")
        for j in range(2 * (i - 1)):
            print("_", end="")
        print("\\")
N = int(input("Enter the number of lines: "))
print_pattern(N)
```

10. Print a number as a 8 segment display N Lines:

---



---



_
_|
|_
_
_|
_|
|_|
|
N=2 N=3 N=4


```python
def print_7_segment_number(number):
    # Segment mapping for each digit
    segments = {
        '0': [' _ ', '| |', '|_|',''],
        '1': ['   ', '  |', '  |'],
        '2': [' _ ', ' _|', '|_ '],
        '3': [' _ ', ' _|', ' _|'],
        '4': ['   ', '|_|', '  |'],
        '5': [' _ ', '|_ ', ' _|'],
        '6': [' _ ', '|_ ', '|_|'],
        '7': [' _ ', '  |', '  |'],
        '8': [' _ ', '|_|', '|_|'],
        '9': [' _ ', '|_|', ' _|']
    }
    lines = ['' for _ in range(3)]
    for digit in number:
        if digit in segments:
            digit_segments = segments[digit]
            for i in range(3):
                lines[i] += digit_segments[i] + ' '
    for line in lines:
        print(line)
number = input("Enter the number: ")
print_7_segment_number(number)
```

11. Print the pattern upto N lines:
1 2
4 3

1 2 3
8 9 4
7 6 5

1 2 3 4
12 13 14 5
11 16 15 6
10 9 8 7

N=2 N=3 N=4

```python
def print_spiral_pattern(N):
    # Initialize variables
    matrix = [[0] * N for _ in range(N)]
    num = 1
    top, bottom, left, right = 0, N - 1, 0, N - 1

    while top <= bottom and left <= right:
        # Fill the top row
        for i in range(left, right + 1):
            matrix[top][i] = num
            num += 1
        top += 1

        # Fill the right column
        for i in range(top, bottom + 1):
            matrix[i][right] = num
            num += 1
        right -= 1

        # Fill the bottom row
        for i in range(right, left - 1, -1):
            matrix[bottom][i] = num
            num += 1
        bottom -= 1

        # Fill the left column
        for i in range(bottom, top - 1, -1):
            matrix[i][left] = num
            num += 1
        left += 1

    # Print the matrix
    for row in matrix:
        print(' '.join(map(str, row)))

# Example usage
N = int(input("N: "))
print_spiral_pattern(N)
```


12. Write a python script that displays the following table
1 1 1 1 1
2 1 2 4 8
3 1 3 9 27
4 1 4 16 64
5 1 5 25 125

```python
def display_table(n):
    for i in range(1, n + 1):
        row = [i]  # Start with the row index
        for j in range(1, 5):
            if j == 1:
                row.append(1)
            elif j == 2:
                row.append(i)
            elif j == 3:
                row.append(i**2)
            elif j == 4:
                row.append(i**3)
        print(' '.join(map(str, row)))

# Example usage
N = int(input("N: "))
display_table(N)
```
