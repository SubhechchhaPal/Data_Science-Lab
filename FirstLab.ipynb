WAP to check whether a number is a palindrome or not

```python
num=int (input("Enter a num: "))
rev=0
temp=num
while(num>1):
  r=int (num%10)
  num=num/10
  rev=(rev*10)+r
if(temp==rev):
  print(rev ,"is Palindrome")
else:
  print(temp ,"is not Palindrome")
```

Write a Java program for following grading system. Note: Percentage>=90% : Grade A Percentage>=80% : Grade B Percentage>=70% : Grade C Percentage>=60% : Grade D Percentage>=40% : Grade E Percentage <40%:Grade F

```python
percent = float(input("Enter Percentage :: "))
if percent >= 90:
    grade = 'A'
elif percent >= 80:
    grade = 'B'
elif percent >= 70:
    grade = 'C'
elif percent >= 60:
    grade = 'D'
elif percent >= 40:
    grade = 'E'
else:
    grade = 'F'
print("Grade ::", grade)
```
